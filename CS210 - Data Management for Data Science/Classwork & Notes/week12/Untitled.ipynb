```python
df.dropna(how='all', subset=['Price', 'Category'], inplace=True)
df
```

```python
for i in df.index:
    if(not isinstance(df.loc[i, 'Category'], str)):
        if (df[df.Category == 'MR']['Price'].min() <= df.loc[i, 'Price'] <= df[df.Category == 'MR']['Price'].max()):
            df.loc[i, 'Category'] = 'MR'
        elif (df[df.Category == 'EC']['Price'].min() <= df.loc[i, 'Price'] <= df[df.Category == 'EC']['Price'].max()):
            df.loc[i, 'Category'] = 'EC'
        elif (df[df.Category == 'LX']['Price'].min() <= df.loc[i, 'Price'] <= df[df.Category == 'LX']['Price'].max()):
            df.loc[i, 'Category'] = 'LX'
        elif (df[df.Category == 'HP']['Price'].min() <= df.loc[i, 'Price'] <= df[df.Category == 'HP']['Price'].max()):
            df.loc[i, 'Category'] = 'HP'
        else :
            continue
```
